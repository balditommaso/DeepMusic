```python
from miditok import REMI, CPWord
from miditoolkit import MidiFile

class MidiVectorizer:
    
    def __init__(self, multi_note=False, pitch_range=None, nb_velocities=None, additional_tokens=None, pad=False, sos_eos=False, mask=False) -> None:
        if multi_note:
            self.tokenizer = CPWord(pitch_range=pitch_range, nb_velocities=nb_velocities, additional_tokens=additional_tokens, pad=pad, sos_eos=sos_eos, mask=mask)
        else:
            self.tokenizer = REMI(pitch_range=pitch_range, nb_velocities=nb_velocities, additional_tokens=additional_tokens, pad=pad, sos_eos=sos_eos, mask=mask)
    
    def encode(self, path):
        midi = MidiFile(path)
        return self.tokenizer.midi_to_tokens(midi)
    
    def decode(self, tokens):
        return self.tokenizer.tokens_to_midi(tokens)
    
```

```python
pitch_range = range(21, 109)
nb_velocities = 32
additional_tokens = {'Chord': False, 'Rest': True, 'Tempo': False, 'Program': False, 'TimeSignature': False,
                     'rest_range': (2, 8),  # (half, 8 beats)
                     'nb_tempos': 64,  # nb of tempo bins
                     }  # (min, max)

vectorizer_remi = MidiVectorizer(pitch_range=pitch_range, nb_velocities=nb_velocities, additional_tokens=additional_tokens, pad=True, sos_eos=True)
vectorizer_cpword = MidiVectorizer(multi_note=True, pitch_range=pitch_range, nb_velocities=nb_velocities, additional_tokens=additional_tokens, pad=True, sos_eos=True)

file = os.path.join(dir_path, os.listdir(dir_path)[34])
midi = midi_util.MidiUtils(file)
print("Original:")
midi.play_midi()

# REMI
tokens_remi = vectorizer_remi.encode(file)
back_to_midi = vectorizer_remi.decode(tokens_remi)
back_to_midi.dump("../decode_example.mid")
midi = midi_util.MidiUtils("../decode_example.mid")
print("Decoded REMI:")
midi.play_midi()

# CPWord
tokens_cpword = vectorizer_cpword.encode(file)
back_to_midi = vectorizer_cpword.decode(tokens_cpword)
back_to_midi.dump("../decode_example.mid")
midi = midi_util.MidiUtils("../decode_example.mid")
print("Decoded CPWord:")
midi.play_midi()
```
